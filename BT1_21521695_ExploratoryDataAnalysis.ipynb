# Chia dataset thành tập train và tập test

```python
from sklearn.model_selection import train_test_split
```

```python
penguins
```

```python
penguins = penguins.dropna()
```

```python
X = penguins.drop(columns = 'sex', axis = 1)
```

```python
X
```

```python
X = X.drop(columns = ['species', 'island'])
```

```python
y = penguins['sex']
```

```python
X = X.to_numpy()
y = y.to_numpy()
```

```python
print(X.shape)
print(y.shape)
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle=True)
```

```python
X_train
```

```python
y_train
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
y = lb.fit_transform(penguins["sex"])
y
```

```python
X_train
```

# Train model với max_depth = None

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test.shape
```

```python
y_pred.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

# Train model với max_depth = 3

```python
model1 = DecisionTreeClassifier(max_depth = 3)
model1.fit(X_train, y_train)
y1_pred  = model1.predict(X_test)
accuracy_score(y_test, y1_pred)
```

# Train model với max_depth = 4

```python
model2 = DecisionTreeClassifier(random_state=0, max_depth = 4)
model2.fit(X_train, y_train)
y2_pred  = model2.predict(X_test)
accuracy_score(y_test, y2_pred)
```

# Train model với max_depth = 7

```python
model3 = DecisionTreeClassifier(random_state=0, max_depth = 7)
model3.fit(X_train, y_train)
y3_pred  = model3.predict(X_test)
accuracy_score(y_test, y3_pred)
```

# Nhận xét

Tăng max_depth thì độ chính xác ngày càng tăng. Trong quá trình thử nghiệm với nhiều max_depth khác nhau thì em nhận thấy tại max_depth = 7 thì độ chính xác có giá trị cao hơn khi max_depth = None. Do vậy, max_depth = None không phải lúc nào cũng cho performance tốt nhất trong tất cả các trường hợp.
